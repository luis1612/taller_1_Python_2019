```python
import pandas as pd
import sqlalchemy
import matplotlib.pyplot as plt 
```

```python
#CONEXION BASE DE DATOS
engine = sqlalchemy.create_engine('mysql+pymysql://root:@localhost:3306/grival')
```

```python

articulo = pd.read_sql_table('articulo',engine)
```

```python

articulo = articulo.dropna(subset=['stock'])
```

```python
articulo.stock = articulo.stock.fillna(0)
```

```python
articulo = articulo.drop_duplicates(subset=['nombre'])
```

```python
# LIMPIA
articulo
```

```python

articulos = articulo[0:5]
nombre = articulo.nombre
cantidad = articulo.stock

plt.figure(figsize=(10,6))
plt.bar(nombre, cantidad, width=0.5, color="yellow")
plt.xticks(nombre)
plt.ylabel('Cantidad')
plt.xlabel('Articulos')
plt.title('DIAGRAMA DE BARRAS DEL STOCK POR ARTICULOS')

```

```python
plt.show()
```

```python
 
articulos = articulo[5:20]
nombre = articulos.nombre
cantidad = articulos.stock

plt.figure(figsize=(10,6))
plt.barh(nombre, cantidad, color="grey")

plt.ylabel('Articulos')
plt.xlabel('Cantidad')
plt.title('EstadÃ­sticas Cantidad Barras horizontales')
plt.show()
```
